```python
import pandas as pd
import numpy as np
```

```python
# This dataset is not included in github and can be found here: 
# https://www.kaggle.com/datasets/fabioturazzi/cryptocurrency-tweets-with-sentiment-analysis
tweets = pd.read_csv("tweets_sentiment.csv")
```

```python
tweets.dtypes
```

```python
# eliminate unnecessary fields
tweets_filtered = tweets[['date', 'mentions', 'replies_count', 'retweets_count', 'likes_count', \
                          'hashtags', 'cashtags', 'neg', 'neu', 'pos', 'compound']]
```

```python
tweets_filtered.sample(10)
```

```python
len(tweets_filtered)
```

```python
bitcoin_tweets = tweets_filtered[tweets_filtered["hashtags"].str.contains("bitcoin|btc", regex=True)]
```

```python
len(bitcoin_tweets)
```

```python
bitcoin_tweets.head()
```

```python
bitcoin_tweets["compound_times_replies"] = bitcoin_tweets["compound"] * bitcoin_tweets["replies_count"]
bitcoin_tweets["compound_times_retweets"] = bitcoin_tweets["compound"] * bitcoin_tweets["retweets_count"]
bitcoin_tweets["compound_times_likes"] = bitcoin_tweets["compound"] * bitcoin_tweets["likes_count"]
```

```python
bitcoin_tweets.head()
```

```python
bitcoin_pivot = bitcoin_tweets.pivot_table(index="date", \
                                           values=["cashtags", "replies_count", "retweets_count",\
                                                   "likes_count", "neg", "neu", "pos", "compound", \
                                                  "compound_times_replies","compound_times_likes", \
                                                   "compound_times_retweets"],
                                          aggfunc={"cashtags" : "count", "replies_count" : np.sum,\
                                                   "retweets_count": np.sum, "likes_count": np.sum,\
                                                   "neg": np.mean, "neu": np.mean, "pos": np.mean,\
                                                   "compound": np.mean,  "compound_times_replies": np.sum,\
                                                   "compound_times_likes": np.sum, "compound_times_retweets": np.sum})
```

```python
bitcoin_pivot.rename(columns={'cashtags': "count"}, inplace=True)
bitcoin_pivot["compound_weighted_replies"] = bitcoin_pivot["compound_times_replies"] / bitcoin_pivot['replies_count']
bitcoin_pivot["compound_weighted_likes"] = bitcoin_pivot["compound_times_likes"] / bitcoin_pivot['likes_count']
bitcoin_pivot["compound_weighted_retweets"] = bitcoin_pivot["compound_times_retweets"] / bitcoin_pivot['retweets_count']
```

```python
bitcoin_pivot.sample(10)
```

```python
bitcoin_pivot[['count', 'compound', 'neg', 'neu', 'pos', 'retweets_count', 'likes_count', 'replies_count',\
               'compound_weighted_replies', 'compound_weighted_likes', 'compound_weighted_retweets']]\
.to_csv("../data/bitcoin_daily_tweets.csv")
```
