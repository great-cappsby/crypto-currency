```python
import pandas as pd
import numpy as np
```

```python
dogecoin = pd.read_csv("../data/doge_daily_modified.csv")
```

```python
dogecoin.head()
```

```python
dogecoin = dogecoin.drop(columns=['Unnamed: 0'])
dogecoin.head()
```

```python
# Create a daily change percent
dogecoin['Daily_Change_Perc'] = dogecoin["Daily_Change"] / dogecoin["Open"]
```

```python
# Re-add gold
gold = pd.read_csv("../data/gold_modified.csv")
gold['Daily_Change'] = gold["Close/Last"] - gold["Open"]
gold['Daily_Change_Perc'] = gold["Daily_Change"] / gold["Open"]
gold["Increased"] = np.where(gold["Daily_Change"] > 0, 1, 0)
# Get date in right format for merge
gold["Date"] = pd.to_datetime(gold["Date"]).dt.strftime('%Y-%m-%d')
gold_needed_columns = gold[["Date", "Close/Last", "Open", "Daily_Change", "Daily_Change_Perc", "Increased"]]
```

```python
# Merge Gold Columns Back in
dogecoin = dogecoin.merge(gold_needed_columns, on="Date", suffixes=("", "_Gold"))
```

```python
# Re-add sp500
sp500 = pd.read_csv("../data/sp_500.csv")
sp500['Daily_Change'] = sp500["Close/Last"] - sp500["Open"]
sp500['Daily_Change_Perc'] = sp500["Daily_Change"] / sp500["Open"]
sp500["Increased"] = np.where(sp500["Daily_Change"] > 0, 1, 0)
# Get date in right format for merge
sp500["Date"] = pd.to_datetime(sp500["Date"]).dt.strftime('%Y-%m-%d')
sp_needed_columns = sp500[["Date", "Close/Last", "Open", "Daily_Change", "Daily_Change_Perc", "Increased"]]
```

```python
dogecoin = dogecoin.merge(sp_needed_columns, on="Date", suffixes=("", "_SP500"))
```

```python
dogecoin.head()
```

```python
dogecoin = dogecoin.rename(columns={'Close/Last': 'Close/Last_Gold', 'Increased': "Increased_Gold"})
```

```python
# set y label
dogecoin["label"] = dogecoin["Daily_Change_Ind"].shift(1)
```

```python
tweets = pd.read_csv('../data/dogecoin_daily_tweets.csv')
```

```python
tweets.head()
```

```python
tweets = tweets.fillna(0)
tweets = tweets.rename(columns={'date': 'Date'})
```

```python
dogecoin = pd.merge(dogecoin, tweets, on="Date", how='left', suffixes=("", ""))
```

```python
dogecoin.head()
```

```python
for (columnName, columnData) in dogecoin.iteritems():
    print('Column Name : ', columnName)
    print('Any NAs : ', columnData.isnull().any())
    try:
        print('Any INFs:', np.isinf(columnData).any())
    except:
        print("Couldn't check")
```

```python
dogecoin['compound_weighted_replies'] = dogecoin['compound_weighted_replies'].fillna(0)
dogecoin['compound_weighted_retweets'] = dogecoin['compound_weighted_retweets'].fillna(0)
dogecoin['compound_weighted_likes'] = dogecoin['compound_weighted_likes'].fillna(0)
dogecoin['compound'] = dogecoin['compound'].fillna(0)
dogecoin['retweets_count'] = dogecoin['retweets_count'].fillna(0)
dogecoin['likes_count'] = dogecoin['likes_count'].fillna(0)
dogecoin['replies_count'] = dogecoin['replies_count'].fillna(0)
dogecoin['neg'] = dogecoin['neg'].fillna(0)
dogecoin['pos'] = dogecoin['pos'].fillna(0)
dogecoin['neu'] = dogecoin['neu'].fillna(0)
dogecoin['count'] = dogecoin['count'].fillna(0)
```

```python
print(len(dogecoin))
dogecoin = dogecoin.dropna()
print(len(dogecoin))
```

```python
trends = pd.read_csv("../data/dogecoin_google_trends.csv")
```

```python
trends = trends.drop(columns=['isPartial']).rename(columns={'dogecoin': 'goog_trend', 'date': 'Date'})
trends.head()
```

```python
dogecoin = dogecoin.merge(trends, on='Date')
```

```python
dogecoin.head()
```

```python
# Train - Test Split
from sklearn.model_selection import train_test_split
random_state = 12345

train_set, test_set = train_test_split(dogecoin, test_size=0.2, random_state=random_state)
```
